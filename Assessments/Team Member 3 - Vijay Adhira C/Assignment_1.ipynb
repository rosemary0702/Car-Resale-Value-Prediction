## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
df=pd.DataFrame()
df['time']= pd.date_range(start="1/1/2023", end="2/10/2023", freq='24H')
df['year']= df['time'].dt.year
df['month']= df['time'].dt.day
df['day']= df['time'].dt.day
df.head(len(df["time"]))
```
